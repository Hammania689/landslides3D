```python
def draw_before_after(after='after_west.ply', before='before_west.ply'):
    ## Basic Visualization of before and after models in the same viewer
    after_pcd = read_point_cloud(after)
    after_pcd.paint_uniform_color([1, 0.706, 0])
    
    cloud = PointCloud()
#     cloud.points = Vector3dVector(np.asarray(after_pcd.points) + np.array([-100, -20, -100]))
    cloud.points = Vector3dVector(np.asarray(after_pcd.points) + np.array([.1, 0, 0]))
    
    #     cloud.paint_uniform_color([1, 0.706, 0])
    cloud.paint_uniform_color([0, 0.651, 0.914])
    
    
    before_pcd = read_point_cloud(before)
    before_pcd.paint_uniform_color([0, 0.651, 0.914])
    
    ## Draws Geometeries with respected color parameters
    draw_geometries([after_pcd, before_pcd])
```

```python
# draw_geometries([after])
draw_before_after()
poincloud = read_point_cloud('after_west.ply')
print(np.asarray(poincloud.points)[0])
print(np.asarray(poincloud.points) + np.array([0, 20, 0]))
cloud = PointCloud()
cloud.points = Vector3dVector(np.asarray(poincloud.points) + np.array([0, 20, 0]))
print('\nTest\n',np.asarray(cloud.points))
# draw_geometries([cloud])
```
